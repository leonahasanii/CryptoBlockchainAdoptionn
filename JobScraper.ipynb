```python
columns_to_drop = [
    "company_industry",
    "company_url",
    "company_logo",
    "company_url_direct",
    "company_addresses",
    "company_num_employees",
    "emails",
    "listing_type",
    "job_function",
    "job_level",
    "currency",
    "max_amount",
    "min_amount",
    "interval",
    "salary_source",
    "job_type", 
    "job_url_direct",
    "company_revenue",
    "company_description",
    "skills",
    "experience_range",
    "company_rating",
    "company_reviews_count",
    "vacancy_count",
    "work_from_home_type",
    "id"
]

df_filtered = df.drop(columns=columns_to_drop, errors="ignore")
```

```python
df_filtered["is_remote"] = df_filtered["is_remote"].fillna(False)
```

```python
df_filtered = df_filtered.dropna(subset=["company"])
```

```python
duplicate_rows = df_filtered[df_filtered.duplicated()]
print(f"Number of duplicated rows: {len(duplicate_rows)}")
```

```python
df_filtered = df_filtered.drop_duplicates()
```

```python
data =df_filtered.copy()
```

```python
# Count the frequency of each site
site_counts = data["site"].value_counts().reset_index()
site_counts.columns = ["Site", "Job Postings"]

# Here we create a bar chart 
fig = px.bar(
    site_counts,
    x="Site",
    y="Job Postings",
    text="Job Postings",
    title="Job Postings by Source Website",
    color_discrete_sequence=["#1f77b4"] 
)

# Update layout 
fig.update_layout(
    title_font=dict(size=20, family="Times New Roman"),
    font=dict(size=14, family="Times New Roman"),
    xaxis_title="Source Website",
    yaxis_title="Number of Job Postings",
    plot_bgcolor="white",
    showlegend=False,
    margin=dict(l=40, r=40, t=80, b=40)
)

# Add subtle gridlines
fig.update_xaxes(showgrid=False)
fig.update_yaxes(showgrid=True, gridcolor="lightgray")

# Show the chart
fig.show()
```

We can see from the chart above that most of the jobs that were scraped from 5 of the jobs platforms, most of the blockchain and the cryptocurrency jobs were from linkedin and indeed. However, we should take into consideration that most of the jobs that were scraped from LinkedIn do not show the complete list of the jobs posted from LinkedIn, because the platforms itself has limits which does not allow to webscrape until a certain reach of the jobs postings.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
import numpy as np

# Create a labeled dataset from the earlier manually classified companies
labeled_data = {
    'company': list(industry_mapping.keys()),
    'industry': list(industry_mapping.values())
}
df_labeled = pd.DataFrame(labeled_data)

# Use TF-IDF to vectorize company names
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(df_labeled['company'])

# Fit a KNN classifier on the labeled data
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X, df_labeled['industry'])

# Transform all company names from the original dataset
X_all = vectorizer.transform(data['company'].fillna(''))

# Predict industries
data['industry_predicted'] = knn.predict(X_all)

# Count predicted industry frequencies
predicted_counts = data['industry_predicted'].value_counts().reset_index()
predicted_counts.columns = ['Industry', 'Count']

predicted_counts = pd.DataFrame(predicted_counts)

predicted_counts
```
